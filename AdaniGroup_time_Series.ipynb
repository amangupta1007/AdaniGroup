# Time Reshampling

```python
df.resample(rule='A').min()
```

```python
df.resample(rule='A').min().plot(figsize=(16,9))
```

```python
df.resample(rule='A').max()
```

```python
df.resample(rule='A').max().plot(figsize=(16,9))
```

```python
df.resample(rule='BQS').min().plot(figsize=(16,9))
```

```python
df.resample(rule='BQS').max().plot(figsize=(16,9))
```

```python
df['Open'].resample(rule='M').max().plot(figsize=(16,9), kind='bar')
```

```python
df['Open'].resample(rule='M').min().plot(figsize=(16,9), kind='bar')
```

```python
df['Open'].resample(rule='A').max().plot(figsize=(16,9), kind='bar')
```

```python
df['Open'].resample(rule='BA').max().plot(figsize=(16,9), kind='bar')
```

```python
df['Open'].resample(rule='BQS').max().plot(figsize=(16,9), kind='bar')
```
